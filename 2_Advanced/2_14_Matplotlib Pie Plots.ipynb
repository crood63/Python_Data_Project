```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## ðŸŸ© Health Insurance Pie Chart (2.14.1)

```python
health_df = df.copy()

health_df['job_health_insurance'].value_counts().plot(kind='pie', startangle=90, autopct='%1.1f%%', title='Offered Health Insurance')
plt.ylabel('')
plt.show()
```

---

## ðŸŸ¨ Job Schedule Distribution (2.14.2)

```python
sch_df = df.copy()

main_five_sch_list = list(sch_df['job_schedule_type'].value_counts().iloc[3:].index)

other_sch_df = sch_df[sch_df['job_schedule_type'].isin(main_five_sch_list)]
other_sch_df['job_schedule_type'] = 'Other'
other_sch_df = other_sch_df['job_schedule_type'].value_counts()

main_df = sch_df['job_schedule_type'].value_counts().head(3)

main_df = pd.concat([main_df, other_sch_df]).sort_values()

main_df.plot(kind='pie', startangle=343, autopct='%1.1f%%', title='Job Schedule Distribution', rotatelabels=True)
plt.ylabel('')
plt.tight_layout()
plt.show()
```

## ðŸŸ¥ Degree Mentioned Pie Chart (2.14.3)

```python
degree_df = df.copy()

degree_df = degree_df['job_no_degree_mention'].value_counts()
degree_df.plot(kind='pie', startangle=90, autopct='%1.1f%%', title='Degree Mentioned')
plt.ylabel('')
plt.show()
```
