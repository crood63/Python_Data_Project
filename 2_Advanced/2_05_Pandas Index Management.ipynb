```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Set Index and Display Rows (2.5.1)

```python
df.set_index('job_posted_date', inplace=True)
df
```

---

## ðŸŸ© Reset Index and Drop Existing (2.5.2)

```python
df.reset_index(inplace=True)
df
```

---

## ðŸŸ© Sort Index (2.5.3)

```python
df.sort_index(inplace=True)
df
```

---

## ðŸŸ¨ Pivot Table Reset & Set Index (2.5.4)

```python
pivot_data = df.pivot_table(index='job_country', columns='job_title_short', values='salary_year_avg', aggfunc=['median'])
pivot_data.reset_index(inplace=True)
pivot_data.set_index('job_country', inplace=True)
pivot_data
```
