```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

## ðŸŸ© Data Scientist Hourly Salaries (2.17.1)

```python
ds_hourly_df = df[df['job_title_short'] == 'Data Scientist'].copy()

ds_hourly_df['salary_hour_avg'].plot(kind='hist', bins=30, edgecolor='black', color='green')

ax = plt.gca()

ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x)}'))

plt.xlim(0, 150)
plt.title('Data Scientist Hourly Salaries')
plt.show()
```

---

## ðŸŸ¨ Job Postings by Number of Skills (2.17.2)

```python
skills_df = df.copy()

skills_df = skills_df.dropna(subset=['job_skills'])

skills_df['job_skill_num'] = skills_df['job_skills'].apply(lambda x: len(x))

skills_df['job_skill_num'].plot(kind='hist', bins=50, edgecolor='black', color='yellow')

plt.xlim(0, 20)
plt.title('Job Postings by Number of Skills')
plt.xlabel('Numebr of Skills')
```

---

## ðŸŸ¥ Data Engineer Salaries (2.17.3)

```python
de_sal_df = df[df['job_title_short'] == 'Data Engineer'].copy()

de_sal_df['salary_year_avg'].plot(kind='hist', bins=30, edgecolor='black', color='red')

plt.xlim(0, 400000)

ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))

plt.title('Data Engineer Salary Distribution')
```
