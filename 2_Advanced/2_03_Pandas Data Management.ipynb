```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Drop Missing Values (2.3.1)

```python
df_drop_nan = df.copy()
df_drop_nan = df_drop_nan.dropna(subset=['salary_rate', 'job_skills', 'job_location', 'job_schedule_type', 'job_via'])
df_drop_nan.count()
```

## ðŸŸ¨ Fill Missing Salaries (2.3.3)

```python
df_salary_fill = df_drop_nan.copy()
yearly_salary_med = df_salary_fill['salary_year_avg'].median()
hourly_salary_med = df_salary_fill['salary_hour_avg'].median()

df_salary_fill['salary_year_avg'] = df_salary_fill['salary_year_avg'].fillna(yearly_salary_med)
df_salary_fill['salary_hour_avg'] = df_salary_fill['salary_hour_avg'].fillna(hourly_salary_med)

df_salary_fill
```
