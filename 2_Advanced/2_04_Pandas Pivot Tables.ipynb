```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Count Job Postings by Title and Country (2.4.1)

```python
df.pivot_table(index='job_country', values='job_title_short', aggfunc='count')
```

---

## ðŸŸ¨ Salary Stats by Job Title (2.4.2)

```python
df.pivot_table(index='job_title_short', values='salary_year_avg', aggfunc=['mean', 'median', 'max', 'min'])
```

---

## ðŸŸ¥ Sum Salary by Title and Country (2.4.3)

```python
df.pivot_table(index='job_country', columns='job_title_short', values='salary_year_avg', aggfunc='sum')
```
