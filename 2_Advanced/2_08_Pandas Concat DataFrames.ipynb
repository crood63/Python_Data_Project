```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## ðŸŸ© Concat Job Postings (2.8.1)

```python
lower_income = df[df['salary_year_avg'] < 100000]
lower_income
```

```python
higher_income = df[df['salary_year_avg'] >= 100000]
higher_income
```

```python
allIncomes = pd.concat([higher_income, lower_income], ignore_index=True)
allIncomes
```

---

## ðŸŸ¨ Concat Monthly Postings (2.8.2)

```python
monthly_df = df.copy()
monthly_df['month_posted'] = monthly_df['job_posted_date'].dt.strftime('%b')
month_list = monthly_df['month_posted'].unique()

monthly_df_dict = {month: monthly_df[monthly_df['month_posted'] == month] for month in month_list}

first_quarter_df = pd.concat([monthly_df_dict['Jan'], monthly_df_dict['Feb'], monthly_df_dict['Mar']], ignore_index=True)
first_quarter_df['month_posted'].value_counts().plot(kind='bar')
```

## ðŸŸ¨ Concat Job Types (2.8.3)

```python
job_type_df = df.copy()

jobs = job_type_df['job_title_short'].unique()

job_type_dict = {job: df[df['job_title_short'] == job] for job in jobs}

data_job_df = pd.concat([job_type_dict['Senior Data Engineer'], job_type_dict['Data Analyst'], job_type_dict['Data Engineer'], job_type_dict['Data Scientist'], job_type_dict['Senior Data Analyst'], job_type_dict['Senior Data Scientist']], ignore_index=True)

data_job_df['job_title_short'].value_counts().plot(kind='bar')
```
