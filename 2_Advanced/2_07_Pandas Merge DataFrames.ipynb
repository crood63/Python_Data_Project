```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

## ðŸŸ© Merge US Jobs and Monthly Postings (2.7.1)

```python
us_jobs = df[df['job_country'] == 'United States']
us_jobs
```

```python
monthly_postings = df.copy()
monthly_postings['month_posted'] = monthly_postings['job_posted_date'].dt.month_name()
monthly_postings
```

```python
us_jobs.merge(monthly_postings)
```

---

## ðŸŸ¨ Merge Non-Remote Jobs and Salary (2.7.2)

```python
non_remote = df[df['job_work_from_home'] == False]
non_remote.reset_index(inplace=True)
non_remote.drop('index', axis=1, inplace=True)
non_remote
```

```python
higher_salary = df[df['salary_year_avg'] >= 100000]
higher_salary.reset_index(inplace=True)
higher_salary.drop('index', axis=1, inplace=True)
higher_salary
```

```python
non_remote.merge(higher_salary, on=['job_title', 'company_name'])
```

---

## ðŸŸ¨ Merge Company Salary and Remote Count (2.7.3)

```python
company_sal = df.pivot_table(index='company_name', values='salary_year_avg', aggfunc='median')
company_sal
```

```python
remote_count = df[df['job_work_from_home'] == True]
remote_count = remote_count[['job_title_short', 'job_location', 'salary_year_avg', 'company_name']]
remote_count.set_index('company_name')
```

```python
company_sal.merge(remote_count, on='company_name')
```
